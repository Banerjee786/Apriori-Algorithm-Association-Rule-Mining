```python
# Importing Libraries
import pandas as pd
import csv
```

### Task 1: DataSet Preprocess
Input is dataset_group.csv
Preprocessing the given CSV to remove the date of transactions and grouping the Transaction Ids'
Exporting the modified dataset in a file named 'dataset.csv'.


```python
# # # The following piece of code reads the given CSV and groups the Items as per Transaction Id # # #

data = pd.read_csv('dataset_group.csv')   # Reading dataset_group csv
new_data = data.groupby(['Trans_Id'])['Item'].apply(list)  # Grouping the Transaction Ids' to get the Items
new_data.to_csv('sample_dataset.csv')  # Converting the grouped items into a csv dataset 
```

```python
# # #  Solution  # # #
# # # The following piece of code creates the processed dataset ready for Apriori Algorithm # # #

with open('sample_dataset.csv', 'r') as f:
    reader = csv.reader(f)        # Reading sample_dataset csv (where Items are grouped as per Transaction Id)
    my_list = list(reader)        # Creating a list of all Items
r = 0
for m in my_list:
    str1 = my_list[r][1].replace("[","")  # Parsing the list's items
    str2 = str1.replace("'","")           # Parsing the list's items
    str3 = str2.replace("]","")           # Parsing the list's items
    new_list = str3.split(",")            # Parsing the list's items to create the Final Item List
    r=r+1
    with open("dataset.csv", "a", newline='') as fp:  # Creating & Appending the processed dataset.csv
        wr = csv.writer(fp, dialect='excel')          # Initializing write to a csv also readable in excel format
        wr.writerow(new_list)                         # Writing each row to the csv       
     
        
```

### Task 2: Run apriory.py and Evaluate Results 

The Apriori Algorithm is executed "3" times for different combinations of support and confidence.
Printing the results of apriory for 'dataset.csv' using the methods described in apriori.py


```python
print ('# # # # # # # # # Case:1 # # # # # # # # #') 
print ('Case 1 (minimum support=0.15 and minimum confidence=0.6 (60%))')
print ('\nCase 1 Reasoning:')
print ('''Here Minimum Support is chosen as 0.15 (low value w.r.t the support of the dataset) and Confidence as 0.6 or 60%   (low Confidence w.r.t the confidence of the association rules)
Support is the indication of how frequently the itemset appears in the dataset
Confidence is an indication of how often the rule holds true
A low Min. Support value would produce more itemsets which satisfies (> Min. Support) while low confidence
would result in more association rules for the given low Confidence.
Hence, depending on the given dataset and the business needs, a tradeoff between Min. Support and Confidence
is to be decided, which holds true in discovering the interestingness of the measures.
In this case, a high min. support of 0.15 ensures that all 1-itemset and 2-itemset have individual support > 0.15,
while the final association rules have individual confidence > Min. Confidence of 60%
\nThe numerical values at the end of each itemset corresponds to the Support of the itemset satisfying Min. Support  of 0.15
The numerical values at the end of each rule corresponds to the Confidence of the Association Rule greater (satisfying) than the specified benchmarked Confidence of 60%''')
print ('\nCase 1 Output:')
%run -i apriori.py -f dataset.csv -s 0.15 -c 0.6   # Magic line command to run the apriori.py script computing apriori algorithm with Min. Support as 0.15 and Confidence as 60%
print('''As per our rules with Min. Confidence of 60%,
Rule # 1 says a person buying Toilet Paper has 74.5% chance of buying Vegetable
Last Rule says a person buying dinner rolls and poultry together has 84% chance of buying vegetables''')
```

```python
print ('# # # # # # # # # Case:2 # # # # # # # # #') 
print ('Case 2 (minimum support=0.3 and minimum confidence=0.7)') 
print ('Case 2 Reasoning:')
print ('''Here Minimum Support is chosen as 0.3 (high value w.r.t the support of the dataset) and Confidence as 0.7 or 70%   (decent Confidence w.r.t the confidence of the association rules)
Support is the indication of how frequently the itemset appears in the dataset
Confidence is an indication of how often the rule holds true
A high Min. Support value would produce less itemsets which satisfies (> Min. Support), while high confidence
would result in less association rules for the given Confidence.
In this case, a high min. support of 0.3 ensures that all 1-itemset and 2-itemset have individual support > 0.3,
while the final association rules have individual confidence > Min. Confidence of 70%
Hence, depending on the given dataset and the business needs, a tradeoff between Min. Support and Confidence
is to be decided, which holds true in discovering the interestingness of the measures.
\nThe numerical values at the end of each itemset corresponds to the Support of the itemset satisfying Min. Support  of 0.3
The numerical values at the end of each rule corresponds to the Confidence of the Association Rule greater (satisfying) than the specified benchmarked Confidence of 70%''')
print ('\nCase 2 Output:') 
%run -i apriori.py -f dataset.csv -s 0.3 -c 0.7   # Magic line command to run the apriori.py script computing apriori algorithm with Min. Support as 0.3 and Confidence as 70%
print('''As per our rules with Min. Confidence of 70%,
Rule # 1 says a person buying Poultry has 78.5% chance of buying Vegetables
Rule # 2 says a person buying Yogurt has 81.8% chance of buying vegetables
Rule # 3 says a person buying Eggs has 82.5% chance of buying vegetables''')
```

```python
print ('# # # # # # # # # Case:3 # # # # # # # # #') 
print ('Case 3 (minimum support=0.3 and minimum confidence=0.8)')
print ('Case 3 Reasoning:')

print ('''Here Minimum Support is chosen as 0.3 (high value w.r.t the support of the dataset) and Confidence as 0.8 or 80%   (optimistic Confidence w.r.t the confidence of the association rules)
Support is the indication of how frequently the itemset appears in the dataset
Confidence is an indication of how often the rule holds true
A high Min. Support value would produce less itemsets which satisfies (> Min. Support), while high confidence
would result in less association rules for the given Confidence, but better and more confirm solid Association Rules
In this case, a high min. support of 0.3 ensures that all 1-itemset and 2-itemset have individual support > 0.3,
while the final association rules have individual confidence > Min. Confidence of 80%
Hence, depending on the given dataset and the business needs, a tradeoff between Min. Support and Confidence
is to be decided, which holds true in discovering the interestingness of the measures.
\nThe numerical values at the end of each itemset corresponds to the Support of the itemset satisfying Min. Support  of 0.3
The numerical values at the end of each rule corresponds to the Confidence of the Association Rule greater (satisfying) than the specified benchmarked Confidence of 80%''')
print ('\nCase 3 Output:')
%run -i apriori.py -f dataset.csv -s 0.3 -c 0.8   # Magic line command to run the apriori.py script computing apriori algorithm with Min. Support as 0.3 and Confidence as 80%
print('''As per the above generated rules,
A person buying Yogurt has 81.8% chance of buying Vegetables
A person buying Eggs has 82.5% chance of buying Vegetables''')
```

## Comparison & Evaluation of Results :

#### Case 1 : Min. Support = 0.15 and Min. Confidence = 60% yields a lot of itemsets > Min. Support and a lot of Association Rules each having confidence > 60%. Hence, with only a min. confidence of 60% it is unreliable to predict and adopt the association rules.

#### Case 2 : Min. Support = 0.3 and Min. Confidence = 70% yields considerable increase in prediction of Association Rule Mining. The increased Min. Support of 0.3 ensures more frequently appearing itemsets in whole dataset. 

#### Case 3 : Keeping the Min. Support as 0.3, since it seems a perfect probability for this dataset. Min. Support of 0.4 caused high pruning of items and very low itemsets being left for Association Rule Mining. The Min. Confidence of 80% provides a stronghold and increased confidence in the Generated Association Rules and therefore can be adopted for business requirements.
